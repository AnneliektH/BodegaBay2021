## Python notebook to make a bar chart of what PIGEON vOTUs were found in what vegetation groups
- Make bar chart with what PIGEON vOTUs found in what vegetation group
- Make a pie chart of what type of PIGEON vOTUs were found

```python
# imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.mode.chained_assignment = None 
sns.set_theme(style="whitegrid")
from matplotlib.colors import ListedColormap
```

```python
# Read in df with num of vOTUs that were found at what vegetation group
df = pd.read_csv('../../new_pigeon_mapping/percent_pigeon_votus_habitrat.csv')
df.head()
```

```python
# Create a color palette
pal = ['#8c510a','#38afcd','#006994','#bf812d',
      '#bababa', '#800066','#8dd35f']
```

```python
# Draw a nested barplot by vOTUs and habitat group
g = sns.catplot(
    data=df, kind="bar",
    x="salinity_group", y="sal_g_amount", hue="ecosystem",
    palette=pal, alpha=.9, height=6
)
g.despine(left=True)
g.set_axis_labels("", "Number of vOTUs recovered")
g.legend.set_title("")

# Save as pdf
g.savefig("../plots/pigeon_sal_group_barchart.pdf") 
```

```python
# Pie chart for which PIGEON vOTUs were found in the BB dataset

# Hardcoded percentages for each of the categories (ugly sorry)
y = np.array([17, 29, 0, 
              30, 0,13, 4,
              0, 7])
# Set the labels
mylabels = [ "Marine", "Freshwater", "Non-marine Saline and Alkaline", 
            "Agricultural Soil", "Rhizosphere Soil","Natural Soil","Peat soil", 
            "Digestive", "Other"]

# Set color palette
mycolors = [ "#006994",  "#38afcd", "#c7eae5", 
            "#8c510a", "#a6d96a", "#bf812d", "#4b5320",
            "#de77ae", "#bababa" ]

# Plot pie chart
plt.pie(y, labels = mylabels, colors=mycolors)

# Save figure as pdf
plt.legend(title = 'vOTU Categories')
plt.savefig("../plots/PIGEON_found_all.pdf")
plt.show() 
```
